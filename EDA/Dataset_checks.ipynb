```python
import pandas as pd
import numpy as np

train = pd.read_csv('/Users/deannalakshman/Desktop/IIT/YEAR 2/SEM 2/DSPLC/Coursework technical/Hotel Chain A - Supporting Resources/Hotel-A-train.csv')
val   = pd.read_csv('/Users/deannalakshman/Desktop/IIT/YEAR 2/SEM 2/DSPLC/Coursework technical/Hotel Chain A - Supporting Resources/Hotel-A-validation.csv')
test  = pd.read_csv('/Users/deannalakshman/Desktop/IIT/YEAR 2/SEM 2/DSPLC/Coursework technical/Hotel Chain A - Supporting Resources/Hotel-A-test.csv')

print('Train shape     :', train.shape)
print('Validation shape:', val.shape)
print('Test shape      :', test.shape)
```

```python
for name, df in [('TRAIN', train), ('VALIDATION', val), ('TEST', test)]:
    print(f'\n{"="*50}')
    print(f'{name}')
    print(f'{"="*50}')
    print(df.dtypes.to_string())
```

```python
for name, df in [('TRAIN', train), ('VALIDATION', val), ('TEST', test)]:
    print(f'\n{name}')
    missing = df.isnull().sum()
    pct     = (missing / len(df) * 100).round(2)
    summary = pd.DataFrame({'Missing': missing, 'Pct (%)': pct})
    print(summary[summary['Missing'] > 0].to_string() or '  No missing values')
```

```python
print('In TRAIN but not TEST      :', sorted(set(train.columns) - set(test.columns)))
print('In VALIDATION but not TEST :', sorted(set(val.columns)   - set(test.columns)))
print('In TEST but not TRAIN      :', sorted(set(test.columns)  - set(train.columns)))
```

```python
date_cols = ['Expected_checkin', 'Expected_checkout', 'Booking_date']

for name, df in [('TRAIN', train), ('VALIDATION', val), ('TEST', test)]:
    df = df.copy()
    for col in date_cols:
        df[col] = pd.to_datetime(df[col], errors='coerce')

    invalid_dates = df[df['Expected_checkin'] < df['Booking_date']].shape[0]
    invalid_stay  = df[df['Expected_checkout'] <= df['Expected_checkin']].shape[0]
    df['Lead_Time']     = (df['Expected_checkin'] - df['Booking_date']).dt.days
    df['Stay_Duration'] = (df['Expected_checkout'] - df['Expected_checkin']).dt.days

    print(f'\n{name}')
    print(f'  Check-in before booking date : {invalid_dates}')
    print(f'  Checkout before check-in     : {invalid_stay}')
    print(f'  Lead time     — min: {df["Lead_Time"].min()}  max: {df["Lead_Time"].max()}')
    print(f'  Stay duration — min: {df["Stay_Duration"].min()}  max: {df["Stay_Duration"].max()}')
```

```python
print('TRAIN — exact Reservation_Status values:')
for v in train['Reservation_Status'].unique():
    print(f'  repr: {repr(v)}  count: {(train["Reservation_Status"] == v).sum()}')
```

```python
# Check capacity violations
for name, df in [('TRAIN', train), ('VALIDATION', val), ('TEST', test)]:
    violations = df[df['Adults'] + df['Children'] > 5]
    print(f'{name} — capacity violations (Adults + Children > 5): {len(violations)}')
```

```python
# Look at what the violations actually look like
violations_train = train[train['Adults'] + train['Children'] > 5]

print('Adults + Children distribution in violations:')
print((violations_train['Adults'] + violations_train['Children']).value_counts().sort_index())

print('\nAdults range in violations:')
print(violations_train['Adults'].value_counts().sort_index())

print('\nChildren range in violations:')
print(violations_train['Children'].value_counts().sort_index())
```

```python
status_map = {'Check-Out': 1, 'Canceled': 2, 'No-Show': 3}

# After fixing the casing issue first:
for df in [train, val]:
    df['Reservation_Status'] = df['Reservation_Status'].str.strip().replace('Check-out', 'Check-Out')
    df['Reservation_Status_Code'] = df['Reservation_Status'].map(status_map)

# Then your binary targets:
train['Cancelled'] = (train['Reservation_Status'] == 'Canceled').astype(int)
train['No_Show']   = (train['Reservation_Status'] == 'No-Show').astype(int)
```
